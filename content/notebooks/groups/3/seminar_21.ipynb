### Задание 3 - Придумайте как посчитать кол-во лет гроссмейстерам в дату рейтинга

#### Группировка данных

```python
fide_historical.head()
```

```python
fide_historical.groupby('rank').count()
```

```python
fide_historical.groupby('rank')['country'].unique()
```

```python
fide_historical.groupby(['rank', 'country']).count()
```

```python
fide_historical.groupby(['rank', 'country'])['name_title'].unique()
```

```python
fide_historical.groupby('country')['rank'].agg('mean')
```

```python
fide_historical.groupby('country')['rank'].agg(['mean', 'median', 'sum', 'count'])
```

```python
fide_historical.groupby('country')['rank'].agg('mean').sort_values('rank')
```

```python
fide_historical.groupby('country')['rank'].agg('mean').reset_index().sort_values('rank', ascending = False)
```
