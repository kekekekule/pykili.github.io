```python
import csv

def tax(salary, proc = 13):
    for key in salary.keys():
        salary[key] = salary[key] * (1 - proc / 100)
    return(salary)


salary = {}
with open('employee_salary.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile, delimiter = '\t')
    for line in reader:
        salary[line['employeeId']] = salary.get(line['employeeId'], 0) + int(line['value'])
print(tax(salary))
```

```python
if line['employeeId'] in salary:
    salary[line['employeeId']] += int(line['value'])
else:
    salary[line['employeeId']] = int(line['value'])
```

```python
import json
from pprint import pprint
ct = {}
with open("weather_16 (1).json", "r", encoding = "utf-8") as read_file:
    data = json.load(read_file)
    for k in range(100):
        temp = data[k]['main']['temp_max'] - data[k]['main']['temp_min']
        city = data[k]['city']['name']
        c = {city: temp}
        ct.update(c)
        
sort = sorted(ct.items(), key=lambda x: x[1], reverse=True)

for i in sort:
    print(i[0], i[1])
```
